# There are two types of categorical variables
- Nominal : simply have categories
- Ordinal : have categories in whcih rank is important e.g. poor, good, very good etc.

# Handle nominal variables
- one hot encoding
- Mean encoding

# Handle ordinal variables
- Label encoding
- Target guided mean encoding

```python
import pandas as pd
import numpy as np
```

## one hot encoding
-- When we have limited number of categories in a column, we can use one hot encoding where we convert each category into a column

```python
dataset = pd.read_csv(r'C:\Users\52054867\machine learning\titanic.csv')
```

```python
dataset.head()
```

```python
## check distinct counts of few categorical columns
dataset.Sex.unique() # array(['male', 'female'], dtype=object)
dataset.Embarked.unique() # array(['S', 'C', 'Q', nan], dtype=object)
```

```python
df = pd.get_dummies(dataset['Sex'])
```

```python
dataset_new = pd.concat([dataset,df],axis=1)
```

```python
dataset_new
```

```python
dataset_new.drop('Sex', inplace=True, axis=1)
```

```python
dataset_new
```

## Handling column with multiple categories. 

When we have such column, we only consider those values whose frequency is higher than others. 

```python
dataset_new.Cabin.unique()
```

```python
list_top_10 = list(dataset_new.Cabin.value_counts().sort_values(ascending=False).head(10).index)
```

```python
list_top_10
```

```python
for column in list_top_10:
    dataset_new[column] = np.where(dataset_new['Cabin'] == column, 1,0)
```

```python
dataset_new
```

# Mean encoding

- In Mean Encoding we take the number of labels into account along with the target variable to encode the labels into machine comprehensible values
-  [Number of true targets under the label Moscow/ Total Number of targets under the label Moscow ] 
- Mean encoding can embody the target in the label whereas label encoding has no correlation with the target

```python
dataset
```

```python
mean_encode = dataset.groupby(['Embarked'])['Survived'].mean().to_dict()
mean_encode
```

```python
dataset['embarked_mean_encode'] = dataset['Embarked'].map(mean_encode)
```

```python
dataset
```

## label encoding
- Label Encoding refers to converting the labels into numeric form so as to convert it into the machine-readable form

## why we should not use label encoding in nominal columns
Limitation of label Encoding

Label encoding convert the data in machine readable form, but it assigns a unique number(starting from 0) to each class of data. This may lead to the generation of priority issue in training of data sets. A label with high value may be considered to have high priority than a label having lower value.

```python
dataset['Embarked'].value_counts()
```

```python
dataset['Embarked'] = np.where( dataset['Embarked'].isnull(),'S', dataset['Embarked'])
```

```python
dataset['Embarked'].unique()
```

```python
# Add a new column to peform label encoding

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
dataset['Embarked']= label_encoder.fit_transform(dataset['Embarked'])
```
